```python
# Data
x_data = [[0,0],
          [0,1],
          [1,0],
          [1,1]]
y_data = [[0],
          [0],
          [0],
          [1]]
```

```python
# 1. x_data, y_data 를 numpy array 로 변경하여라
# 2. keras model 을 생성해 보세요 sequential 
# 3. node =1, activation ="sigmoid" input_shape=??? 인 Dense layer를 2에서 만든 model 에 추가하세요
# 4. model compile : optimizer = "SGD", loss = binary_crossentropy, metrics="accuracy"
# 5. model fit : batch+size = 4. epochs = 100
# 6. np.round(model.predict(x_data)) 결과 -> [0],[0], [0], [1]
import tensorflow as tf
import numpy as np
tf.random.set_seed(1234)
```

```python
# 딥러닝 model 에 input 하는 data는 전부 숫자형으로 넣어줘야 한다. 
x_data = np.array(x_data)
y_data = np.array(y_data)
```

```python
x_data.shape[1:]
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units = 1, activation = "sigmoid", input_shape=(2,))) # node :1 (출력 값이 1개니깐) input_shape : x_data.shape[1:]
model.compile(optimizer = "SGD", loss = "binary_crossentropy", metrics="accuracy")
```

```python
model.fit(x_data, y_data, batch_size=4, epochs = 500) # batch_size 에 따라느 GD, SGD, mini-batch GD 가 된다. 총 50번 학습
```

```python
np.round(model.predict(x_data)) #반올림을 통해서 0과 1로 분류
```

# 데이터를 나누는 단위
- Epoch : 전체 sample DATA 를 이용하여 한바퀴 학습 1epoch
- Step : Weight 와 Bias를 1회 업데이트 하는걸 1Step 이라고 한다. 
- Batch size : 1step 에서 사용한 data의 수
  - Batch size : 100
  - Step : 5
  - 총 500개의 데이터를 이용해 학습 
- 관계식
  - s = (n * e) / b
  - s : step 의 수
  - n : number of sample 
  - e : epoch
  - b : batch size

# Batch Szie 에 따른 Gradient Descent
- cost function vs loss function
  - loss function 손실을 측정 model 의 성능을 측정
  - cost function : 1/n 을 통해서 학습데이터에 반영 

- cost function : 1/n 로 loss function 의 값을 나눠준다
- n = m : 전체 학습데이터의 갯수 -> Full Batch GD
- n = 1 : 1개를 이용해 학습(랜덤으로 추출) -> SGD
  - Vector 연산의 장점이 사라진다.
  - 가중치가 빠르게 업데이트 된다.
  - 적은 Epoch 으로 학습 가능
- n != m : 보통 $2^n$ 개로 Batch size 지정(메모리가 2진법) -> mini batch GD
  - Vector 연산의 장점은 갖고 가면서 가중치는 빠르게 업데이트 시킬수 있다. 



# callblack 함수
### 1. 학습률(learning rate): ReduceLROnPlateau
- 학습률 동적 조절
- ReduceLROnPlateau는 학습률이 개선되지 않을 때, 학습률을 동적으로 조정하여 학습률을 개선하는 효과를 기대할수 있다.
- 경사하강법에 의하여 학습을 하는 경우 Local Minima에 빠져버리게 되면, 더이상 학습률이 개선되지 않고 정체되거나, 심하게 튀는 현상을 제어할때 사용할수 있다.
- https://deep-deep-deep.tistory.com/56


### 2. weight 중간 저장: ModelCheckpoint
- 모델이 학습하면서 정의한 조건을 만족했을 때 Model의 weight 값을 중간 저장
- 모델이 개선된 validation score를 도출해낼 때마다 weight를 중간 저장함으로써, 혹시 중간에 memory overflow나 crash가 나더라도 다시 weight를 불러와서 학습을 이어나갈수 있게 해준다.


### 3. 조기종료: EarlyStopping

- EarlyStopping 콜백을 활용하면, model의 성능 지표가 설정한 epoch동안 개선되지 않을 때 조기 종료할 수 있습니다.
- EarlyStopping과 이전에 언급한 ModelCheckpoint 콜백의 조합을 통하여, 개선되지 않는 학습에 대한 조기 종료를 실행하고, ModelCheckpoint로 부터 가장 best model을 다시 로드하여 학습을 재게할 수 있습니다.
- https://deep-deep-deep.tistory.com/55


```python
from tensorflow.keras.callbacks import ReduceLROnPlateau, ModelCheckpoint, EarlyStopping
```

```python
help(ReduceLROnPlateau)
```

```python
epoch = 2000
batch_size = 4


filepath = f"model_check_point_epoch_{epoch}_batch_{batch_size}.h5"
reduceLR = ReduceLROnPlateau(monitor="loss",
                             factor = 0.5, # loss 값을 개선하기 위한 학습률 변경 시작 0.1 -> 0.2 0> 0.4 
                             patience = 10) # epoch 10 동안 loss 값이 개선되지 않으면 callback 호출

checkpoint = ModelCheckpoint(filepath = filepath,
                             monitor = "loss",
                             save_best_only = True) #가장 Best 한 값만 저장한다. 

earlystopping = EarlyStopping(monitor = "loss",
                              patience=10)

optimizer = tf.keras.optimizers.SGD(learning_rate = 0.1)
```

```python
model_xor = tf.keras.Sequential()
optimizer = tf.keras.optimizers.SGD(learning_rate = 0.3)

model_xor.add(tf.keras.layers.Dense(units = 2, activation = "sigmoid", input_shape=(2,))) # 출력이 2개
model_xor.add(tf.keras.layers.Dense(units = 1, activation = "sigmoid")) # 출력 1개 
model_xor.compile(optimizer = optimizer, loss = "binary_crossentropy", metrics="accuracy")
```

```python
model_xor.fit(x_data_xor, y_data_xor, batch_size=4, epochs=4000, callbacks=[reduceLR, checkpoint, earlystopping])
```
