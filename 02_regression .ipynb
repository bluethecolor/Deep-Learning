```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_random_data():
    x = np.random.uniform(low=-2, high=2, size=100)
    y = []
    for t in x:
        r = np.random.normal(loc=0.0, scale=(0.5 + t*t/3), size=None)
        y.append(r)
    return  x, x**3 + -1 + np.array(y)
```

```python
x, y =  make_random_data()
```

```python
plt.plot(x, y, 'bo')
plt.show()
```

```python
# 위와같은 Data Set을 학습 시키고 결과를 plot으로 출력해 보세요 
activation_fun = "sigmoid" #tanh, relu ....
units_1 = 1 # 학습시킬 node 의 갯수
units_2 = 1 # y.shape 의 값이 (100,) 이므로 y의 1개의 shape은 1 이다. 
input_shape = (1,) # x.shape 의 값이 (100,) 이므로 y의 1개의 shape은 1 이다. 
model = tf.keras.Sequential([
                             tf.keras.layers.Dense(units=units_1, activation=activation_fun,input_shape=input_shape),
                             tf.keras.layers.Dense(units=units_2)

])
model.compile(optimizer = "sgd", loss= "mse")
model.fit(x, y, epochs = 100, batch_size = 100)
```

```python
models = []
for i in range(1, 5):
  units_1 = i # 학습시킬 node 의 갯수
  activation_fun = ["sigmoid", "tanh", "relu"]
  units_2 = 1 # y.shape 의 값이 (100,) 이므로 y의 1개의 shape은 1 이다. 
  input_shape = (1,) # x.shape 의 값이 (100,) 이므로 y의 1개의 shape은 1 이다. 
  sub_model =[]
  for activation in activation_fun:
    model = tf.keras.Sequential([
                                 tf.keras.layers.Dense(units=units_1, activation=activation, input_shape=input_shape),
                                 tf.keras.layers.Dense(units=units_2)
    ])

    model.compile(optimizer = "sgd", loss= "mse")
    model.fit(x, y, epochs = 100, batch_size = 100)
    sub_model.append(model)
  models.append(sub_model)
```

```python
models
```

```python
for model_act in models:
  for i in len(model_act):
    plt.plot(x, y, 'bo')
    x_predict = np.linspace(min(x), max(x))
    y_predict = model_act[i].predict(x_predict)
    plt.subplot(len(model_act)+1, 1, i+1)
    plt.plot(x_predict, y_predict, "red")
    plt.show()
```

```python
model_list = []
for i in range(1, 10):
    model_regression = tf.keras.Sequential()
    model_regression.add(tf.keras.layers.Dense(i, activation="relu", input_shape=(1,)))
    model_regression.add(tf.keras.layers.Dense(1))
    model_regression.compile(optimizer="sgd", loss="mse")
    model_regression.fit(x,y, epochs=100)
    model_list.append(model_regression)
```

```python
plt.subplots_adjust(bottom=5, top=10, 
                    wspace=0.2, 
                    hspace=0.35)
for i in range(0, len(model_list)):
    x_predict = np.arange(min(x), max(x), 0.01)
    y_predict = model_list[i].predict(x_predict)
    plt.subplot(len(model_list)+1,1, i+1)
    plt.plot(x,y, "bo")
    plt.plot(x_predict, y_predict, "red")
    plt.title(f"units = {i}")
```
